```python

# ‡∏£‡∏ß‡∏°‡πÑ‡∏ß‡πâ‡πÉ‡∏ô dict ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ß‡∏ô‡∏•‡∏π‡∏õ
dfs = {
    "2023": {"room": df_room_2023, "summary": df_summary_2023},
    "2024": {"room": df_room_2024, "summary": df_summary_2024},
    "2025": {"room": df_room_2025, "summary": df_summary_2025}
}

# Function: ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
def clean_data(room_df, summary_df):
    # ‡∏•‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ã‡πâ‡∏≥
    room_df = room_df.drop_duplicates().copy()
    summary_df = summary_df.drop_duplicates().copy()

    # ‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤ missing
    room_df["Guests"] = room_df["Guests"].fillna("Unknown")
    summary_df["Guest names"] = summary_df["Guest names"].fillna("Unknown")

    # ‡πÅ‡∏õ‡∏•‡∏á‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà
    for col in ["Booked At", "Modified At", "Cancelled At", "Check - In", "Check - Out"]:
        if col in room_df.columns:
            room_df[col] = pd.to_datetime(room_df[col], errors="coerce")

    for col in ["Booked-on date", "Check-in", "Check-out"]:
        if col in summary_df.columns:
            summary_df[col] = pd.to_datetime(summary_df[col], errors="coerce")

    # ‡πÅ‡∏õ‡∏•‡∏á‡∏£‡∏≤‡∏Ñ‡∏≤‡∏à‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
    if "Total price" in summary_df.columns:
        summary_df["Total price"] = (
            summary_df["Total price"]
            .replace("THB", "", regex=True)
            .replace(",", "", regex=True)
            .astype(str)
            .str.strip()
            .astype(float)
        )

    return room_df, summary_df

# ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏ó‡∏∏‡∏Å‡∏õ‡∏µ
for year in dfs :
    dfs[year]["room"], dfs[year]["summary"] = clean_data(dfs[year]["room"], dfs[year]["summary"])
```

```python
import pandas as pd

# ‡πÇ‡∏´‡∏•‡∏î‡πÑ‡∏ü‡∏•‡πå
room_2023 = pd.read_csv("2023_reservations_room_stays_report.csv")
summary_2023 = pd.read_csv("2023_reservations_summary_report.csv")
room_2024 = pd.read_csv("2024_reservations_room_stays_report-2.csv")
summary_2024 = pd.read_csv("2024_reservations_summary_report-2.csv")
room_2025 = pd.read_csv("2025_reservations_room_stays_report-3.csv")
summary_2025 = pd.read_csv("2025_reservations_summary_report-3.csv")

# ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏£‡∏ß‡∏° Room + Summary ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏£‡∏∞‡∏ö‡∏∏‡∏õ‡∏µ
def merge_with_year(room, summary, year):
    merged = pd.merge(
        room,
        summary,
        how="outer",
        left_on="Booking Reference" if "Booking Reference" in room.columns else "Booking reference",
        right_on="Booking reference" if "Booking reference" in summary.columns else "Booking Reference",
        suffixes=('_room', '_summary')
    )
    merged["Year"] = year
    return merged

# ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏∏‡∏Å‡∏õ‡∏µ
merged_data = pd.concat([
    merge_with_year(room_2023, summary_2023, 2023),
    merge_with_year(room_2024, summary_2024, 2024),
    merge_with_year(room_2025, summary_2025, 2025)
], ignore_index=True)

# ‡πÅ‡∏õ‡∏•‡∏á‡∏ß‡∏±‡∏ô‡πÄ‡∏ß‡∏•‡∏≤
merged_data["Check-in"] = pd.to_datetime(
    merged_data["Check-in"] if "Check-in" in merged_data.columns else merged_data["Check - In"],
    errors="coerce"
)
merged_data["Check-out"] = pd.to_datetime(
    merged_data["Check-out"] if "Check-out" in merged_data.columns else merged_data["Check - Out"],
    errors="coerce"
)

# ‚úÖ ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏µ‡∏•‡∏î‡πå‡πÉ‡∏´‡∏°‡πà
merged_data["Nights"] = (merged_data["Check-out"] - merged_data["Check-in"]).dt.days
merged_data["Check-in Weekday"] = merged_data["Check-in"].dt.day_name()

def get_season(month):
    if month in [11, 12, 1]: return "Cool"
    elif month in [2, 3, 4, 5]: return "Hot"
    elif month in [6, 7, 8, 9, 10]: return "Rainy"
    return "Unknown"
merged_data["Season"] = merged_data["Check-in"].dt.month.apply(get_season)

def classify_customer(occupancy):
    try:
        adults = int(str(occupancy).split(" - ")[0])
        if adults == 1: return "Solo"
        elif adults == 2: return "Couple"
        elif adults > 2: return "Group"
    except:
        return "Unknown"
merged_data["Customer Type"] = merged_data["Occupancy"].apply(classify_customer)

merged_data["Is Weekend"] = merged_data["Check-in"].dt.weekday >= 5

# üíæ ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏´‡∏°‡πà
merged_data.to_csv("merged_data_with_fields.csv", index=False)
```
